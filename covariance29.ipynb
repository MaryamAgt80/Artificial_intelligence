```python
#29 covariance
import numpy as np
import matplotlib.pyplot as plt
def func_means(values):
    mean_values=np.mean(values)
    return [xi-mean_values for xi in values]

def covariance(valuex,valuey):
    sum=np.dot(func_means(valuex),func_means(valuey))
    sum=sum/(len(valuex)-1)
    return sum

page_speed=np.random.normal(100,10,1000)
page_request=np.random.normal(100,10,1000)
covariance(page_speed,page_request)
plt.scatter(page_speed,page_request)
type(page_request)
```

```python
#30 formool correlation ضریب همبستگی
import scipy.stats as sp
def correlation(x,y):
    var_iance=covariance(x,y)
    return var_iance/x.std()/y.std()


    
```
