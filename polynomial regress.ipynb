```python
#38
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2)
x=np.random.normal(3,1,1000)
y=np.random.normal(50,5,1000)/x
plt.scatter(x,y)
```

```python
x=np.array(x)
y=np.array(y)
polyvalue=np.poly1d(np.polyfit(x,y,5))  #5:degree equation
print(polyvalue)
```

```python
linsp=np.linspace(0.05,6,100) # 0.05:start 6:end 100:softness
plt.scatter(x,y)
plt.plot(linsp,polyvalue(linsp),c='r')
```
