```python
import matplotlib.pyplot as plt
import pandas as pd
cars=pd.read_excel('cars_data.xlsx')
cars.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
import numpy as np
#plt.scatter(cars['Price'],cars['Liter'])
x=cars['Price']
y=cars['Liter']

data=[]
for i in range(len(x)):
    data.append([x[i],y[i]])

model=KMeans(n_clusters=8)
model=model.fit(scale(data))
plt.scatter(x,y,c=model.labels_.astype(np.float))
print(model.labels_)
```
