```python
import pandas as pd
import numpy as np
movies_data=pd.read_csv('userbase_data/movies_data.txt',sep='\t',names=['user_id','movi_id','rating'],usecols=range(3))
user_data=pd.read_csv('userbase_data/movies_item.txt',sep='|',names=['movi_id','title'],usecols=range(2))
user_data=pd.merge(user_data,movies_data)
user_data.head()
```

```python
movirating=user_data.pivot_table(index='user_id',columns='title',values='rating')
movirating.head(10)
```

```python
ToyStoryMovi=movirating['Toy Story (1995)']

print(ToyStoryMovi)
```

```python
similarmovies=movirating.corrwith(ToyStoryMovi)
similarmovies=similarmovies.dropna()
similarmovies=pd.DataFrame(similarmovies,columns=['similarity'])
```

```python
#58 groupby and popular 
moviesstats=user_data.groupby('title').agg({'rating':[np.size,np.mean]})
moviesstats.head()
```

```python
movipopular=moviesstats['rating']['size']>=100

movipopular.head()
#movipopular.columns.values.tolist()
```

```python
moviesstats[movipopular].sort_values([('rating','mean')],ascending=False)
```

```python

moviesfinaly=moviesstats[movipopular].join(pd.DataFrame(similarmovies,columns=['similarity']))
df=moviesfinaly.sort_values(['similarity'],ascending=False)
df.head()
```
