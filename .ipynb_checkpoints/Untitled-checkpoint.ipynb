```python
#23-1
import matplotlib.pyplot as plt
values=[22,30,40,50,80]
colors=['#FF0000','#00FFFF','#FFFF00','#FFA500','#FF00FF']
labels=['iran','russian','america','holand','portghal']
myexpold=[0.2,0,0,0,0.1]
plt.pie(values,labels=labels,colors=colors,explode=myexpold,shadow=True,startangle = 0)
plt.legend(title='colors')
plt.show()
```

```python
#23-2
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["s", "a", "m", "z"])
y = np.array([16, 23, 22, 45])
colors=['#FF0000','#00FFFF','#FFFF00','#FFA500']
plt.bar(x,y,color=colors,width=0.5)
plt.show()
```
