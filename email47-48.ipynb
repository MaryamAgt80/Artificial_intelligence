```python
#47 spamemail1
import os 
import io
import numpy as np
from pandas import DataFrame as DF
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
#OS.WALK SEARCH TO NET
def readfile(pathroot):
    for root, dirs, files in os.walk(pathroot):
        for filename in files:
            path=os.path.join(root,filename)
            file=io.open(path, mode='r',encoding='latin1')
            lines=[]
            is_body=False
            for line in file:
                if is_body:
                    lines.append(line)
                elif line=='\n':
                    is_body=True
            message="\n".join(lines)
            
            yield message,path
                  
def returndataframe(path,classfictaion):
    listmespath=[]
    index=[]
    for message,path in readfile(path):
        listmespath.append({'message':message,'class':classfictaion})
        index.append(path)
    Data_list=DF(listmespath,index=index)
    return Data_list
    
emailsdata=DF({'message':[],'class':[]})   

emailsdata=emailsdata.append(returndataframe('E:\download\spam email folder\emails\email','not_spam'))
emailsdata=emailsdata.append(returndataframe('E:\download\spam email folder\emails\spam','spam'))

emailsdata['class'].values
```
