```python
#32,34
from numpy import random
from numpy import random
person_number={20:0,30:0,40:0,50:0,60:0,70:0}
person_buy={20:0,30:0,40:0,50:0,60:0,70:0}
age_list=[20,30,40,50,60,70]
all_buy=0
for i in range(0,100000):
    age=random.choice(age_list)
    person_number[age]+=1
    do_buy=random.choice([1,0])
    person_buy[age]+=do_buy
    all_buy+=do_buy
    

random.seed(0)
print(random.seed(0))
    
```

```python
#36-37
import matplotlib.pyplot as plt
x=random.normal(3,1,1000)
y=100-(x+random.normal(0,1,1000))*3
import matplotlib.pyplot as plt 
from scipy.stats import linregress
plt.scatter(x,y)
slope,intercept,r_value,p_value,str_error=linregress(x,y)
print(slope,intercept,r_value,p_value,str_error)
def linreg(xi):
    return slope*xi+intercept
fitflip=linreg(x)
plt.plot(x,fitflip,c='r')
#f(x)=slope*x+intercept
#r_value^2:      
```
