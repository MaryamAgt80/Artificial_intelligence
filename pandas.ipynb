```python
#25 pandas
import pandas as pd
import matplotlib.pyplot as plot
data=pd.read_csv('cars.csv')
data.head()
```

```python
counts_gears=data['# Gears'].value_counts()
counts_gears.plot(kind='bar')
counts_gears[8]
```

```python
import seaborn as sb 
sb.set()
counts_gears.plot(kind='bar')
```

```python
sb.distplot(data['CombMPG'])
```

```python
#26 pandas2
data2=data[['Cylinders','HwyMPG','Eng Displ']]
data2.head()
```

```python
sb.pairplot(data2,hue='Cylinders',height=2.5)
```

```python
sb.scatterplot(x='HwyMPG',y='Eng Displ',data=data)
```

```python
sb.jointplot(x='HwyMPG',y='Eng Displ',data=data)
```

```python
#27 linear by seaborn 
sb.lmplot(x='HwyMPG',y='Eng Displ',data=data)
```

```python
 sb.set(rc={'figure.figsize':(15,5)})
box1=sb.boxplot(x='Transmission',y='Eng Displ',data=data)
box1.set_xticklabels(box1.get_xticklabels(),rotation=20)
```

```python
sb.set(rc={'figure.figsize':(15,5)})
box2=sb.swarmplot(x='Transmission',y='Eng Displ',data=data)
box2.set_xticklabels(box2.get_xticklabels(),rotation=20)
```

```python
sb.set(rc={'figure.figsize':(15,5)})
box2=sb.countplot(x='Transmission',data=data)
box2.set_xticklabels(box2.get_xticklabels(),rotation=20)
```
